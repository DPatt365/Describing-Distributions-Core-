```python
mean = df['TotalCharge'].mean()
mean
```

```python
median = df['TotalCharge'].median()
median
```

```python
col='TotalCharge'

feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std

fix,ax = plt.subplots(figsize=(10,6))
sns.histplot(df['TotalCharge'],ax = ax,kde = True)
ax.axvline(mean,color = 'red',label='Mean')
ax.axvline(median,color = 'green',ls='--',label='Median')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder=0)

ax.legend();
```

```python
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std

fix,ax = plt.subplots(figsize=(10,6))
sns.histplot(df['VitD_levels'],ax = ax,kde = True)
ax.axvline(mean,color = 'red',label='Mean')
ax.axvline(median,color = 'green',ls='--',label='Median')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder=0)
```
